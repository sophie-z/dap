```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3>Reading DAP_2022_DogOverview_v1.0.csv</h3> 
The dog overview dataset contains 35 variables, providing information about 43517 dogs (35 columns 43517 rows).

```python
# turns the csv into a DataFrame
dog_overview = pd.read_csv('DAP_2022_DogOverview_v1.0.csv', low_memory=False)

# number of columns
num_columns = dog_overview.shape[1]
print(f"Number of rows: {len(dog_overview)}")
print(f"Number of columns: {num_columns}")

# number of unique values per column
dog_overview.nunique(axis=0)

dog_overview.describe(include='all')
```

Notably, this dataset includes both purebred and mixed breed dogs of 4021 unique breeds.

The most common owner-reported breed is Labrador Retriever, followed by Golden Retriever, German Shepard Dog, Australian Shepherd, and Poodle. The 6th most common breed is Unknown/Unknown — many dogs in this study are undetermined mixed breeds.

```python
# number of unique breeds
unique_breeds = dog_overview['Breed'].unique().tolist()
print(f"Number of unique breeds: {len(unique_breeds)}")

# top 10 most common dog breeds
most_common_breeds = dog_overview['Breed'].value_counts()[:10].index.tolist()
print(f"Top 10 most common breeds: {most_common_breeds}")

# list of each value that appears in the 'Breed' column and its frequency
breed_counts = dog_overview['Breed'].value_counts()

# # bar graph of number of dogs of each breed (too many breeds to be readable...)
# plt.figure(figsize=(20, 8))
# plt.bar(breed_counts.index, breed_counts.values)
# plt.xlabel('Breed')
# plt.xticks(rotation=90, fontsize=6)
# plt.ylabel('Count')
# ax = plt.gca()
# ax.set_ylim([0, 800])
# plt.title('Breeds of Dogs')
# plt.show()
```

There are 17 columns with missing data points, but these missing data points are insignificant for our purposes as they do not provide important identifying information for each dog's condition.

```python
# number of missing data points per column
missing_data = dog_overview.isnull().sum()
missing_data = missing_data[missing_data > 0]
print(missing_data)

# heatmap of missing data points per column
plt.figure(figsize=(20, 10))
sns.heatmap(dog_overview.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Data Pattern")
plt.show()
```

<h3>Reading DAP_2022_HLES_cancer_conditions_v1.0.csv</h3> 
The health condition dataset contains 90 columns and 2762 rows.

```python
# turns the csv into a DataFrame
cancer_conditions = pd.read_csv('DAP_2022_HLES_cancer_conditions_v1.0.csv', low_memory=False)

# number of columns
num_columns = cancer_conditions.shape[1]
print(f"Number of rows: {len(cancer_conditions)}")
print(f"Number of columns: {num_columns}")
```

```python
# number of missing data points per column
missing_data = cancer_conditions.isnull().sum()
missing_data = missing_data[missing_data > 0]
print(missing_data)

# heatmap of missing data points per column
# splitting the DataFrame into three parts so the heatmap is readable
cancer_conditions_1 = cancer_conditions.iloc[:, :30]
cancer_conditions_2 = cancer_conditions.iloc[:, 30:60]
cancer_conditions_3 = cancer_conditions.iloc[:, 60:90]
# creating a figure with three subplots
fig, axes = plt.subplots(3, 1, figsize=(20, 30))  # Adjust the size as needed

# first heatmap
sns.heatmap(cancer_conditions_1.isnull(), cmap='viridis', cbar=False, ax=axes[0])
axes[0].set_title('Heatmap of columns 1-30')
# second heatmap
sns.heatmap(cancer_conditions_2.isnull(), cmap='viridis', cbar=False, ax=axes[1])
axes[1].set_title('Heatmap of columns 31-60')
# third heatmap
sns.heatmap(cancer_conditions_3.isnull(), cmap='viridis', cbar=False, ax=axes[2])
axes[2].set_title('Heatmap of columns 61-90')

plt.tight_layout()
plt.show()
```
